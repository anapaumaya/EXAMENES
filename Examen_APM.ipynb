```python
#Ejercicios 8 y  9
import pandas as pd
import numpy as np

data = pd.DataFrame(np.arange(12).reshape((3, 4)),index=['CDMX', 'Guanajuato', 'Puebla'],columns=['uno', 'dos', 'tres', 'cuatro'])
data
#Lo que explica es que el DataFrame toma argumentos para poder definir los valores y que obtenga la estructura del DataFrame
#El np.arange sirve para que los valores tengan un espacio dentro de la matriz y sirve para poner los números que sean solamente del 1 al 12
#El reshape cambia la figura de la matriz
#El index sirve para poder especificar de que va hacer el indice y mas que nada para que la fila tenga nombre
#Las columnas sirven para especificar el nombre de la columna para el DataFrame
```

```python
data= data.rename(index={'CDMX': 'Ciudad de México', 'Guanajuato' :'Estado de Guanajato', 'Puebla': 'Estado de Puebla' })
data
```
